## 4) How many male billionaires are there compared to the number of female billionares? What percent is that? Do they have a different average wealth?

> **TIP:** The last part uses `groupby`, but the count/percent part does not.
> **TIP:** When I say "average," you can pick what kind of average you use.

```python
#Male billionaires

male = df[df.gender == 'male']
len(male)
```

```python
# Percentage of male billionaires. Answer rounded to two decimal places

round(len(male)/len(df) * 100, 2)
```

```python
# Total number of female billionaires

female = df[df.gender == 'female']
len(female)
```

```python
# Average wealth of male measured in billion dollars and rounded to two decimal places

round(male.networthusbillion.mean(),2)
```

```python
# Average wealth of female measured in billion dollars and rounded to two decimal places

round(female.networthusbillion.mean(),2)
```

```python
round(len(female)/len(df) * 100, 2)
```

## 5) What is the most common source/type of wealth? Is it different between males and females?

> **TIP:** You know how to `groupby` and you know how to count how many times a value is in a column. Can you put them together???
> **TIP:** Use percentages for this, it makes it a lot more readable.

```python
#The count of type of wealth amoung the dataframe

df.groupby(by='sourceofwealth').count().sort_values(by='typeofwealth', ascending=False)
```

```python
#Sources of wealth for females

female_wealth = df[df.gender== 'female']

female_wealth.groupby(by='typeofwealth').count()
```

```python
# Percentage breakdown of sources of wealth for females rounded to two decimal places

round(female_wealth.typeofwealth.value_counts()/len(female_wealth) * 100, 2)
```

```python
#Sources of wealth for males

male_wealth = df[df.gender== 'male']

male_wealth.groupby(by='typeofwealth').count()
```

```python
# Percentage breakdown of sources of wealth for males rounded to two decimal places


round(male_wealth.typeofwealth.value_counts()/len(male_wealth) * 100, 2)
```

## 8) What are the top 10 countries with the most money held by billionaires?

I am **not** asking which country has the most billionaires - this is **total amount of money per country.**

> **TIP:** Think about it in steps - "I want them organized by country," "I want their net worth," "I want to add it all up," and "I want 10 of them." Just chain it all together.

```python
# Top ten countries with most money held by billionaires

df.groupby(by='countrycode').networthusbillion.sum().sort_values(ascending=False).head(10)
```
